```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
```

```python
def normal_distribution(x, y, A=5, sx=5, sy=5, x0=10, y0=10):
    return A * np.exp(-((x-x0)**2/(2*sx**2) + (y - y0)**2/(2*sy**2)))


x_space = np.linspace(0, 20, 21)
y_space = np.linspace(0, 20, 21)
X, Y = np.meshgrid(x_space, y_space)
Z = np.ones((21, 21))

for i,x in enumerate(x_space):
    for j,y in enumerate(y_space):
        Z[i, j] = normal_distribution(x, y)

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X.ravel(), Y.ravel(), Z.ravel())
```
