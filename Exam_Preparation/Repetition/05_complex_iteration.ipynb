```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
from numba import njit
```

```python
def julia_set(c, bounds, max_iterations, resolution):
    x = np.linspace(bounds[0], bounds[1], resolution)
    y = np.linspace(bounds[2], bounds[3], resolution)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y

    img = np.zeros_like(Z, dtype=float)

    for i in range(resolution):
        for j in range(resolution):
            z = Z[i, j]
            for k in range(max_iterations):
                if(abs(z) > 2):
                    break
                z = z**2 + c
            img[i, j] = k

    return img

def mandelbrot_set(bounds, max_iterations, resolution):
    x = np.linspace(bounds[0], bounds[1], resolution)
    y = np.linspace(bounds[2], bounds[3], resolution)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y

    img = np.zeros_like(Z, dtype=float)

    for i in range(resolution):
        for j in range(resolution):
            z = Z[i, j]
            c = Z[i, j]

            for k in range(max_iterations):
                if(abs(z) > 2):
                    break
                z = z**2 + c
            img[i, j] = k

    return img



bounds = [-2, 2, -2, 2]
resolution = 1000
julia_set = julia_set(-0.1 + 0.65j, bounds, 2500, 1000)
mandelbrot_set = mandelbrot_set(bounds, 2500, 1000)


plt.imshow(julia_set, origin='lower')
plt.colorbar()
plt.close()
plt.imshow(mandelbrot_set, origin='lower')
```
