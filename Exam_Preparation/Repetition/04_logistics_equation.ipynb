```python
import numpy as np
import matplotlib.pyplot as plt


def logistic_equation(x, a):
    return a * x * (1 - x)



"""
What we want to do for the logistics equation, is start with a population size and then compute it's growth 
for something like 100 to 200 years aka 100 to 200 iterations, we then plot the end value of the population
"""


a_values = np.linspace(1, 4, 10000, endpoint=False)
starting_populations = np.linspace(0, 1, 10, endpoint=False)[1::]



plt.figure(figsize=(10, 6))
# we want to compute the end value of each starting population for each of the a_values
for population in starting_populations:
    for i in range(1000):
        population = logistic_equation(population, a_values)

    plt.scatter(a_values, population, s=0.5, color='black', alpha=0.1)

plt.savefig('test.png', dpi=500)
```
